```python
from random import randint
def nim(num):
    num = num% 3
    if num==1:
        return 1
    if num == 2:
        return 1
    if num == 0:
        return 2
    
def nim_dumb(num):
    return randint(1,2)
```

```python
def nim_game(agent1, agent2, start = 21, games=10, take=2):

    p1_wins = 0
    for i in range(games):
        print(f"Game {i+1}")
        n = start

        while n >= 1:
            
            # agent 1 turn
            taken = agent1(n)
            n -= taken

            if n == 1:
                print("Player 1 wins")
                p1_wins += 1

                break
                
            # agent 2 turn
            taken = agent2(n)
            n -= taken
            if n <= 1:
                print("Player 2 wins")
                break

    
    print('-'*20)
    print(f"Player 1 wins {p1_wins} out of {games} games")    


nim_game(nim, nim_dumb, 21, 100)
```
