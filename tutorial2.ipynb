Pascal's Triangle is a triangular array of numbers where each number is the sum of the two numbers directly above it. The triangle starts with a single 1 at the top, and each subsequent row has one more number than the row above. The edges of the triangle are always 1s, and each interior number is the sum of the two numbers above it. Pascal's Triangle has applications in algebra, combinatorics, and probability theory. It is used to find coefficients in binomial expansions, among other things. Here's how the first few rows look:

```
    1
   1 1
  1 2 1
 1 3 3 1
1 4 6 4 1
```


Write a function `next_pasc(line)` which takes a line (list of integers) of pascales triangle, and prints the next line.

e.g. `next_pasc([1])`

returns

`[1,1]`

and `next_pasc([1,1])`

returns

`[1,2,1]`
